```python
import matplotlib as mp
from matplotlib import pyplot
```

```python
x = [1,2,3]
y = [1,4,3]

pyplot.plot(x, y)
pyplot.xlabel('X - AXIS')
pyplot.ylabel('Y - AXIS')
pyplot.title('The Demo Graph - The mountain')
pyplot.show()
```

```python
pyplot.figure(figsize=(3,3))

pyplot.show()
```

```python
import numpy as np
a = np.arange(0,10,0.1)
b = np.sin(a)

pyplot.plot(a, b)
pyplot.xlabel('X - AXIS')
pyplot.ylabel('Y - AXIS')
pyplot.title('The Demo Graph - The mountain')
pyplot.show()
```
